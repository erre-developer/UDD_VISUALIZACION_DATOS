```python
##Ruta del GITHUB con los datasets

## https://github.com/msaldiasv/UDD_VISUALIZACION_DATOS/tree/main/data


#Importamos las librerias
import pandas as pd
import glob
import os
```

```python
# Primero especificamos un patrón del archivo y lo pasamos como parámetro en la función glob
all_files = glob.glob('/Users/michelsaldias/Desktop/diplomado_datascience/Visualizacion_de_datos/Trabajo_Final/dataset/prueba_muestraventasanual_*.csv')
```

```python
len(all_files)
```

```python
file_list = []

for f in all_files:
    
    data = pd.read_csv(f, sep=';')
    
    nombre = os.path.basename(f)
    
    data['source_file'] = nombre
    
    file_list.append(data)
    
df = pd.concat(file_list, ignore_index=True)
    
```

```python
##¿Como es el comportamiento por Genero sobre la ulima linea que compraron?


result = df.groupby(['SEXO','LINEA_ULTIMO_PRODUCTO_COMPRDO']).aggregate('count')
display(result)
```
